```python
#pip install html5lib
import bs4
import requests
```

```python
url="https://www.passiton.com/inspirational-quotes"
```

```python
response=requests.get(url)
```

```python
soup=bs4.BeautifulSoup(response.content)
```

```python
soup
```

```python
article_ele=soup.findAll('div',{'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'})
```

```python
url=article_ele[1].img.attrs['src']
response=requests.get(url)
response.content
```

```python
for i,article in enumerate(article_ele):
    with open('inspir{}.jpg'.format(i),'wb') as file:
        img_url=article.img.attrs['src']
        response=requests.get(img_url)
        file.write(response.content)
```
